### 4.1 Date 游늱

La "Date" nos muestra por lo general el d칤a, mes y a침o. Debido a que el a침o ya lo vamos a estudiar en la siguiente variable, y el d칤a exacto no resulta relevante y en muchos de los casos no se espec칤fica, seleccionaremos y tomaremos unicamente el valor de los meses.

```python
data4.Date.value_counts()
```

Como podemos observar los meses tienen la tendencia a ser escritos con tres palabras, con el objetivo de extraer solo el mes de la fecha estableceremos un patr칩n regex definiendo dicha tendencia.

```python
date_patron = r"(\b[a-zA-Z]{3}\b)"
data4.Date = data4.Date.str.extract(date_patron,expand = False)
#data4.Date = data4.Date.astype(str)
```

```python
data4.Date.unique()
```

Una vez obtenido obtenido los meses, procedemos a limpiar las strings que a pesar de cumplir con el patr칩n su valor no corresponde con un mes y a sustituir los NaNs por "UNKNOWN"

```python
data4.Date = data4.Date.replace(["Day","and","day","war","the","mid","few",None],"UNKNOWN")
#data4.Date.unique()
data4
```
